```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

df = pd.read_csv("weather.csv")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['precipitation','temp_max','temp_min','wind']],df.weather,test_size=0.2)
X_train
```

```python
model = keras.Sequential([
    keras.layers.Dense(4, input_shape=(4,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])

# 4 çünkü output 4 sınftan oluşuyor
model.add(keras.layers.Dense(256, activation='sigmoid'))
model.add(keras.layers.Dense(4, activation='sigmoid'))

# 1-0 arasında değer üretiyor
# tek sınıf çıktısı
#model.compile(optimizer='adam',
#              loss='binary_crossentropy',
#              metrics=['accuracy'])

# sigorta yapıp yapmayacağını yüzdesel yakınlık olarak gösteriyor. MNIST gibi
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=5000)
```

```python
predict = model.predict(X_test)
```

```python
for i in range(0, 10):
    print(np.argmax(predict[i]))
```

```python
model.evaluate(X_test, y_test)
```
