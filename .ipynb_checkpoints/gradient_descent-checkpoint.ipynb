```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

df = pd.read_csv("insurance_data.csv")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2)
X_train
```

```python
y_train
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
model = keras.Sequential([
    keras.layers.Dense(2, input_shape=(2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])

model.add(keras.layers.Dense(2, activation='relu'))
# 3 te yapabilirsin. Fakat bu durumda training iin 3 sınıfın olması gerekiyor
#model.add(keras.layers.Dense(2, activation='sigmoid'))

# 1-0 arasında değer üretiyor
# tek sınıf çıktısı
#model.compile(optimizer='adam',
#              loss='binary_crossentropy',
#              metrics=['accuracy'])

# sigorta yapıp yapmayacağını yüzdesel yakınlık olarak gösteriyor. MNIST gibi
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=50)
```

```python
# 6 satırlık test veri seti içerisinde %50 yi aşan sigorta yaptırıyor demek
model.predict(X_test_scaled)
```

```python
# 1 olanlar sigorta yaptırmış.
# yukarısı ile karşılaştırılınca 27 0 yanlış tahmin olmuş
# geri kalanı doğru tahmin
y_test
```

```python
model.evaluate(X_test_scaled, y_test)
```
